```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing  import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("Datasets/train.csv")
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.sample(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
X_train.head(2)
```

```python
#Applying Imputation
si_age= SimpleImputer()
X_train_age= si_age.fit_transform(X_train[['Age']])
X_test_age= si_age.fit_transform(X_test[['Age']])
```

```python
si_embarked= SimpleImputer(strategy='most_frequent')
X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])
X_test_embarked= si_embarked.fit_transform(X_test[['Embarked']])
```

```python
X_train_embarked
```

```python
df['Embarked'].value_counts()
```

```python
#one hot encoding 
ohe_sex= OneHotEncoder(handle_unknown='ignore',sparse=False)
X_train_sex= ohe_sex.fit_transform(X_train[['Sex']])
X_test_sex= ohe_sex.fit_transform(X_test[['Sex']])
```

```python
ohe_embarked= OneHotEncoder(handle_unknown='ignore',sparse=False)
X_train_embarked= ohe_embarked.fit_transform(X_train_embarked)
X_test_embarked= ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
X_train_rem= X_train.drop(columns=['Sex','Embarked','Age'])
X_test_rem= X_test.drop(columns=['Sex','Embarked','Age'])
```

```python
X_train_transformed = np.concatenate((X_train_rem, X_train_sex, X_train_embarked, X_train_age),axis=1)
X_test_transformed = np.concatenate((X_test_rem, X_test_sex, X_test_embarked, X_test_age),axis=1)
```

```python
X_test_transformed.shape
```

```python
clf= DecisionTreeClassifier()
```

```python
clf.fit(X_train_transformed,y_train)
```

```python
y_pred=clf.predict(X_test_transformed)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
